```python
import tensorflow as tf
import numpy as np
from tqdm import tqdm
import matplotlib as plt
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Read a comma-separated values into DataFrame.
iris_data = pd.read_csv('data/iris.csv')
print(iris_data)
```

```python
#Extracting data inputs
all_inputs = iris_data[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']].values
```

```python
#Exctracting types of flower as output
all_types = iris_data['variety'].values
```

```python
#Convert output strings to integers 
#np.unique returns the sorted unique elements of an array. In this case, the indices of the unique array that reconstruct the input array
#tf.keras.utils.to_categorical converts a class vector (integers) to binary class matrix

outputs_values, output_ints = np.unique(all_types, return_inverse = True)
output_categories = tf.keras.utils.to_categorical(output_ints)
```

```python
#Split the data into training and testing subsets by creating an array of data set indexes in random order.
#Then split the index array into two subsets.
#np.random.permutation randomly permutes a sequence
#np.array_split splits array into two subarrays

indexes = np.random.permutation(len(all_inputs))
train_indexes, test_indexes = np.array_split(indexes, 2)
```

```python
#Filling the training and testing arrays by randomly shuffling the data sets
input_train, output_train = all_inputs[train_indexes], output_categories[train_indexes]
input_test, output_test = all_inputs[test_indexes], output_categories[test_indexes]
```

```python
#Create the neural network model

#Helpful link: https://www.tensorflow.org/api_docs/python/tf/keras/layers/Dense

#kr.models.Sequential groups a linear stack of layers into tf.keras.model
#model.add adds hidden layers
#tf.nn.softmax computes softmax activations
#tf.nn.sigmoid computes sigmoid of x element-wise
#Adam is a replacement optimization algorithm for stochastic gradient descent 
#binary_crossentropy computes the categorical cross-entropy loss between true labels and predicted labels.

#model = tf.keras.models.Sequential([
    
    #tf.keras.layers.Dense(16, input_shape = (4,)),
    #tf.keras.layers.Dense(32, activation = tf.nn.softmax), #Hidden Layer
    #tf.keras.layers.Dense(3, activation = tf.nn.sigmoid) #Output Layer
    
#])

model = kr.models.Sequential()
model.add(kr.layers.Dense(16, input_shape=(4,)))
model.add(kr.layers.Activation("sigmoid"))
model.add(kr.layers.Dense(3))
model.add(kr.layers.Activation("softmax"))

model.compile(optimizer = "adam", loss = "categorical_crossentropy", metrics = ["accuracy"])
```

```python
#Model is trained using training data

model.fit(input_train, output_train, epochs = 100, batch_size = 1, verbose = 1)
```

```python
loss, accuracy = model.evaluate(input_test, output_test, verbose=0)

print("\n\nLoss: %10.5f\tAccuracy: %2.2f %%" % (loss, accuracy*100))
```
