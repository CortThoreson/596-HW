```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Create the data set
#tf.constant creates a constant tensor from a tensor-like object.
x_train = tf.constant([[1,0],[0,1],[1,1],[0,0]], 'float32')
y_train = tf.constant([[1],[1],[0],[0]], 'float32')
```

```python
#Helpful link: https://www.tensorflow.org/api_docs/python/tf/keras/Model
#tf.keras is an API
#tf.keras.models groups layers into an object with training and inference features.
#tf.keras.model.Sequential groups a linear stack of layers into a tf.keras.models
#tf.keras.layers flattens the multi-dimensional input tensors into a single dimension, so you can model your input layer and build your neural network model, then pass those data into every single neuron of the model effectively.
#tf.keras.layers.Dense is a densely connect neural network layer
#tf.nn.relu computes rectified linear
#tf.nn.sigmoid computes sigmoid of x element-wise
#Adam is a replacement optimization algorithm for stochastic gradient descent 
#binary_crossentropy computes the cross-entropy loss between true labels and predicted labels.

model = tf.keras.models.Sequential([
    
    tf.keras.layers.Flatten(input_shape = (2,)),
    tf.keras.layers.Dense(32, activation = tf.nn.relu), #Hidden Layer
    tf.keras.layers.Dense(1, activation = tf.nn.sigmoid) #Output Layer
    
])

model.compile(optimizer = "Adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
model.fit(x_train, y_train, epochs = 1000)
```

```python
y_test = tf.constant([[1],[1],[0],[0]], "float32")
model.evaluate(x_train, y_test)
```

```python
print(model.predict(x_train))
```
